## Exercise 1
A well established and diffused method for detecting a disease in blood fails to detect the presence of disease in 15% of the patients that actually have the disease.

A young UniPD startUp has developed an innovative method of screening. During the qualification
phase, a random sample of n = 75 patients known to have the disease is screened using the new
method.

### 1.1 What is the probability distribution of y, the number of times the new method fails to detect the disease?

The numbers of failures of a method follows a Binomial distribution.

### 1.2 On the n = 75 patients sample, the new method fails to detect the disease in y = 6 cases. What is the frequentist estimator of the failure probability of the new method?

The frequentist estimator for the binomial distribution is:
$$p_F= \frac{y}{n} $$

$$\begin{cases}E[p_F]&=p\\ 
    Var[p_F]&=\frac{pq}{n}\\    
    MSE[p_F]&= \frac{p(1-p)}{n}
\end{cases}$$

```r
y <- 6
n <- 75

p_estimator <- y/n
```

```r
cat(paste("The frequentist estimator of the failure probability is ", p_estimator,sep=""))
```

### 1.3 Setup a bayesian computation of the posterior probability, assuming a beta distribution with mean value 0.15 and standard deviation 0.14. Plot the posterior distribution for y, and mark on the plot the mean value and variance

```r
mu <- 0.15
var <- 0.14^2 

betaparams <- function(mu,var)
    {
       return(c( - (mu*(var+(mu^2) - mu))/(var), ( ((var+(mu^2)-mu)*(mu-1))/var )) )
    }

a <- betaparams(mu,var)[1]
b <- betaparams(mu,var)[2]
```

```r
p.res <- 0.001

p <- seq(p.res,1, p.res)

beta_prior <- dbeta(p, a, b)

p.likelihood <- dbinom(x = y, size = n, prob = p)
beta_posterior <- (p.likelihood * beta_prior)

# beta_posterior is yet not noramlized, we compute the normalization factor
norm <- sum(beta_posterior)*p.res

# Then we divide
beta_posterior <- beta_posterior/norm
```

```r
# Check normalization
sum(beta_posterior*p.res)
```

```r
plot(p,beta_posterior,main='Bayesian Prior and Posterior',pch='.',col='red',xlim=c(0,.23),lwd=3,ylab='')
lines(p,beta_posterior,col='red',lwd=2)
lines(p,beta_prior,col='pink',lty=2,lwd=2)
legend(0.17,13,legend=c('posterior','prior'),col=c('red','pink'),lty=c(1,2),lwd=2)
```

```r
ap <- a + y
bp <- b + n - y

bayesianmean <- ap / (ap + bp)
bayesianvar  <- ((1/(n+2))^2)*y*(1-y/n)
```

```r
plot(p,beta_posterior,main='Posterior',pch='.',col='red',xlim=c(0,.23),ylab='')
lines(p,beta_posterior,col='red',lwd=2)
abline(v=bayesianmean)
```

```r
cat(paste("The mean value obtained by the distribution is "
          ,round(bayesianmean,4)," with variance ",round(bayesianvar,4),"\n",sep=""))
```

### 1.4 Perform a test of hypothesis assuming that if the probability of failing to the detect the desease in ill patients is greater or equal than 15%, the new test is no better that the traditional method. Test the sample at a 5% level of significance in the Bayesian way.

```r
p.old <- 15/100
alpha <- 5/100
```

We wish to test $H_0:p\leq p.old$ versus $H_1:p>p.old$ at a level of significance of 5%

```r
i <- 1
area <- 0
while(i<p.old/p.res)
    {
        area <- area + beta_posterior[i]*p.res
        i <- i + 1
    }

if(area>alpha)
    {cat(paste("The result ",round(area*100,2),"% is higher than ",alpha*100,
               "%, therefore we cannot reject the null hypothesis at the ",
               alpha*100,"% level of significance",sep=""))
    } else
        {
            cat(paste("The result ",round(area*100,2),"% is lower than ",alpha*100,
               "%, therefore reject the null hypothesis at the ",
               alpha*100,"% level of significance",sep=""))
        }
```

### 1.5 Perform the same hypothesis test in the classical frequentist way.

```r
# With binom.test() we can compute the p-value
binom.test(y, n, p = p.old,
           alternative = "less",
           conf.level = 0.95)
```

### 2.1 Assuming a uniform prior, compute and plot the posterior distribution for λ, the death rate over the measurement time. Determine the posterior mean, median and variance, and compute the 95% credibility interval.

```r
mu.res <- 0.01
mu <- seq(0,50,mu.res)
```

```r
numericalmode <- function(dist,x)
    {
        res <- x[2] - x[1]
        indx <- match(max(dist),dist)
        return (res*indx)
    }

numericalmean <- function(dist,x)
    {
        res <- x[2] - x[1]
        return(sum(x*dist*res))
    }

numericalmedian <- function(dist,x)
    {
        i <- 1
        area <- 0
        res <- x[2] - x[1]
    
        while(area < 0.5)
            {
              area <- 0
              j <- 1
              while(j < i)  
                  {
                     area <- area + (res * dist[j])   
                     j <- j + 1
                  }
              i <- i + 1
            }
        return(i * res)
    }

numericalvariance <- function(dist,x)
    {
        res <- x[2] - x[1]
        mean <- numericalmean(dist,x)
        var <- sum((x * x * dist * res))
        return(var - mean^2)
    }


getcredibilityinterval <- function(dist,x,conf)
    {
        res <- x[2] - x[1]
        indx <- match(max(dist),dist)
        area <- max(dist)*res
        
        i <- 1
        j <- 1
        while(area < conf)
            {
                if (indx - j > 0 )
                    {
                        area <- area + ( dist[indx+i] + dist[indx-j] ) * res
                        i <- i + 1
                        j <- j + 1
                    }
                # This may solve some problems for distribution where the max is close to 0
                else
                    {
                        area <- area + ( dist[indx+i] ) * res
                        i <- i + 1
                    }
            }
    
        return( c( x[indx-j] , x[indx+i] ) ) 
    }
```

```r
n1.post.unif <- dgamma(mu,sum(y.deaths*n1)+1,length(n1))
n2.post.unif <- dgamma(mu,sum(y.deaths*n2)+1,length(n2))

n1.mean.unif <- numericalmean(n1.post.unif,mu)
n2.mean.unif <- numericalmean(n2.post.unif,mu)

n1.mode.unif <- numericalmode(n1.post.unif,mu)
n2.mode.unif <- numericalmode(n2.post.unif,mu)

n1.median.unif <- numericalmedian(n1.post.unif,mu)
n2.median.unif <- numericalmedian(n2.post.unif,mu)

n1.variance.unif <- numericalvariance(n1.post.unif,mu)
n2.variance.unif <- numericalvariance(n2.post.unif,mu)

n1.credint.unif <- getcredibilityinterval(n1.post.unif,mu,.95)
n2.credint.unif <- getcredibilityinterval(n2.post.unif,mu,.95)

plot(mu,n1.post.unif,pch='.',xlim=c(10,45),ylab='P(mu)',
     main='mu probability distributions and credibility intervals (95%) of the two corps')
lines(mu,n1.post.unif,col=n1.col,lwd=3)
lines(mu,n2.post.unif,col=n2.col,lwd=3)

abline(v=n1.mode.unif,lty=2,col=n1.col,lwd=2)
abline(v=n2.mode.unif,lty=2,col=n2.col,lwd=2)
abline(v=n1.credint.unif[1],lty=2,col=n1.col)
abline(v=n1.credint.unif[2],lty=2,col=n1.col)
abline(v=n2.credint.unif[1],lty=2,col=n2.col)
abline(v=n2.credint.unif[2],lty=2,col=n2.col)

legend(38, .20, legend=c('n1', 'n2'),
       col=c(n1.col, n2.col), lwd=3, cex=1.2)

cat('-------------------------\nASSUMING A UNIFORM PRIOR:\n-------------------------\n')
cat(paste('n1:\n'))
cat(paste('   mean:                ',n1.mean.unif,'\n'))
cat(paste('   median:              ',n1.median.unif,'\n'))
cat(paste('   mode:                ',n1.mode.unif,'\n'))
cat(paste('   variance:            ',n1.variance.unif,'\n'))
cat(paste('   credibility interval: [',n1.credint.unif[1],',',n1.credint.unif[2],']\n'))
cat(paste('\nn2:\n'))
cat(paste('   mean:                ',n2.mean.unif,'\n'))
cat(paste('   median:              ',n2.median.unif,'\n'))
cat(paste('   mode:                ',n2.mode.unif,'\n'))
cat(paste('   variance:            ',n2.variance.unif,'\n'))
cat(paste('   credibility interval: [',n2.credint.unif[1],',',n2.credint.unif[2],']\n'))
```

A study on water quality of streams, a high level of bacter X was defined as a level greater than 100

### 3.1 Find the frequentist estimator for p

```r
y.bad <- 11
n.streams <- 116

p_badstr <- y.bad/n.streams

cat(paste('The frequentist estimator for p is',round(p_badstr,4)))
```

### 3.2 Using a Beta(1, 10) prior for p, calculate and posterior distribution P(p|y)

```r
# We are assuming p can range from 0 to 1 with a probability of Beta(1,10)
p.res = 0.00001
p <- seq(p.res,.3,p.res)
```

```r
ex3betaprior <- function(p)
    {
        return(dbeta(p,1,10))
    }

ex3likelihood <- function(p)
    {
        return(dbinom(x=y.bad,size=n.streams,prob=p))
    }

computeposterior <- function(likelihood, prior)
    {
        
        likelihoodvec <- likelihood(p)
        priorvec <- prior(p)
    
        normalization <- p.res * sum(likelihoodvec*priorvec)
        posterior <- likelihoodvec*priorvec/normalization
    
        return(posterior)
    }
```

```r
ex3post.beta <- computeposterior(ex3likelihood,ex3betaprior)

plot(p,ex3post.beta,pch='.',xlim=c(0,0.3),main='Posterior distribution from beta prior Beta(1,10)')
lines(p,ex3post.beta,col='orange',lwd=2)
lines(p,ex3betaprior(p),col='gold',lwd=2,lty=2)
legend(0.195, 16, legend=c('prior', 'posterior'),
       col=c('orange', 'gold'), lwd=3, cex=1.2, lty=c(1,2))
```

### 3.4 Test the hypotesis
$$ H_o:p=0.1\space\space\space\text{versus}\space\space\space H_1:p\neq0.1 $$
### at 5% level of significance with both the frequentist and bayesian approach

#### Bayesian

```r
idxmean <- 1

while(idxmean < ex3post.beta.mean/p.res)
    {
        idxmean <- idxmean + 1
    }

area <- ex3post.beta[idxmean]*p.res
i <- 0
while(area < 0.95)
    {
        i <- i + 1
        if(idxmean>i)
            {
                area <- area + (ex3post.beta[idxmean-i] + ex3post.beta[idxmean+i])*p.res
            }
        else
            {
                area <- area + (ex3post.beta[idxmean+i])*p.res
            }
    }

credint.baye <- c((idxmean-i)*p.res,(idxmean+i)*p.res)

cat(paste('95% bayesian credibility interval: [',credint.baye[1],',',credint.baye[2],']'))
```
