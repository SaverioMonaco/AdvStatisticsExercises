### Exercise 1

The triangular distribution, in the interval (a, b), is given by the following:

$$f(X)=\begin{cases}\displaystyle\frac{2(x-a)}{(b-a)(c-a)} & a\leq x < c\\\displaystyle \frac{2(b-x)}{(b-a)(b-c)}& c\leq x < b\\\displaystyle  0 & \text{otherwise}\end{cases}$$
where $c\in[a,b]$

[Grafico desmos](https://www.desmos.com/calculator/zyzud8547u)

<img src="https://i.ibb.co/qdJPM8g/Cam-Scanner-04-14-2021-10-32-1.jpg" width="400"/>

We can check the normalization without integrating:

$$\begin{align}&A_1 = (c-a)\frac{2}{(b-a)}\frac{1}{2} = \frac{c-a}{b-a}\\
&A_2 = (b-c)\frac{2}{(b-a)}\frac{1}{2} = \frac{b-c}{b-a}\\ &A = A_1+A_2 = \frac{c-a}{b-a} + \frac{b-c}{b-a} = \frac{1}{b-a}\left(c-a + b-c\right) = \frac{b-a}{b-a} = 1\end{align}$$

So for every $c\in[a,b],\text{ } a<b$ the total area is 1 thus representing a probability distribution.

#### 1.1 Plot the function given the interval (a,b)

```r
triang <- function(x,a,b,c)
    {
        ifelse(a<=x & x<b, ifelse(a<=x & x<c, (2*(x-a))/((b-a)*(c-a)), (2*(b-x))/((b-a)*(b-c)) ), 0)
    }
```

```r
# a = 0 # c = 1 # b = 2 #

options(repr.plot.width=10, repr.plot.height=5)
curve(triang(x,0,2,1), from=-1, to=3, , xlab="x", ylab="y", col='forestgreen', lwd = 3)
```

#### 1.2 Write an algorithm to generate random number from the triangular distribution

To generate random number from the angular distribution I choosed the inverse sampling method:
1. Compute the cumulative distribution function F(X) from the pdf f(x)
2. Compute the inverse of the cumulative distribution function:
$$F(X)=p \to G(p) = X $$


3. Generate random probabilities (from U[0,1]) and use them as an input of the function G(p)

```r
# I computed the functions analitically, although it would have been also possible to use the 
# built in integrate function

ctriang <- function(x,a,b,c)
    {
        ifelse(x<a, 0 ,ifelse( x<=c, ((x-a)^2)/((b-a)*(c-a)) , ifelse( x<b, ( c-a + ((x-c)*(2*b-x-c))/(b-c) )/(b-a) , 1 ) ) )
    }
```

```r
# Here I plotted a triangular pdf and his cumulative distribution function

options(repr.plot.width=10, repr.plot.height=5)
curve(triang(x,0,2,.5), from=-1, to=3, , xlab="x", ylab="y", col='gold', lwd = 3)
curve(ctriang(x,0,2,.5), from=-1, to=3, , xlab="x", ylab="y", col='orange', lwd = 3, add = TRUE)
```

```r
# Now we need the quantile function (the inverse of the cumulative function)

qtriang <- function(p,a,b,c)
    {
        temp <- 2*b*c - c*c + (b - c)*((b-a)*p + a - c)
        ifelse(p<=((c-a)/(b-a)), a + sqrt((b-a)*(c-a)*p) , (2*b - sqrt(4*b*b - 4*temp))/(2) )
    
    }
```

```r
# The function that generates random numbers according to the triangular distribution is just the quantile 
# function for N uniform (0,1) random numbers

rtriang <- function(N,a,b,c)
    {
        qtriang(runif(N),a,b,c)
    }
```

### Exercise 2 - Markov's inequality

#### Markov's inequality represents an upper bound to probability distributions:
$$P(X\geq k)\leq \frac{E[X]}{k}\qquad \text{for $k>0$}$$
#### having a defined function
$$G(k)=1-F(k)\equiv P(X\geq k)$$
#### plot G(k) and the Markov's upper bound for

#### 2.1 The exponential, Exp(λ=1), distribution function

```r
# We need the two functions of the inequality:

# The complementary cumulative function
gexp <- function(k)
    {
        1 - pexp(k,1)
    }

# E[X]/k for k > 0
markovineq <- function(k,mean)
    {
        ifelse(k>0 , mean/k, NaN)
    }
```

```r
# We plot E[X]/k as a red line and the complementary cumulative function as a green line.
# for every X it must be that the y value of the red line is above the y value of the green line

curve(markovineq(x,gexp(.5)),ylim=c(0,6),col='red',lwd=2)
curve(gexp(x),add = TRUE,col='forestgreen',lwd=2)
```

### Exercise 3 - Chebyshev's inequality

#### Chebyshev's inequality tell us that
$$P(|X-\mu|\geq k\sigma)\leq\frac{1}{k^2}$$
#### which can also be written as
$$P(|X-\mu|<k\sigma)\geq 1 - \frac{1}{k^2}$$
#### use R to show, with a plot, that Chebyshev's inequality is an upper bound to the following distributions:

```r
# This function integrates the pdf in the range [mu - k*sigma, mu + k*sigma]
# This is the left-hand side term
cheb <- function(k,mu,s,f)
    {
        ifelse(k>0,integrate(function(x) {f(x)}, lower=(mu-k*s), upper=(mu+k*s))$value,0)
    }
```

```r
# To make a discovery one must obtain that the result obtained by chance must be over 5-sigmas, the probability
# of that can be evaluated using this function:
cheb(5,0,1,dnorm)
```

```r
# The right-hand side term:
rhs <- function(x)
    {
        1 - 1/(x*x)
    }
```

#### 3.1 A normal distribution, N(μ=3,σ=5)

```r
mean <- 3
variance <- 5
sdnorm <- function(x)
    {
       dnorm(x,mean,variance) 
    }
```

```r
# As in exercise 2, we must check that the red line is always above (>=) of the green line:

curve(rhs(x),xlim=c(0,5),col='forestgreen',ylim=c(0,1),lwd=2)
ndx <- seq(0,6,0.1)
temp<-purrr::map(ndx,cheb, mu=mean, s=variance, f=sdnorm)
lines(ndx,temp,col='red',lwd=2)
```

#### The data show how visitors to (or time spent in) categorized places change compared to baseline days. A baseline day represents a normal value for that day of the week. The baseline day is the median value from the 5-week period Jan 3 – Feb 6, 2020. To make the reports useful, categories have been used to group some of the places with similar characteristics for purposes of social distancing guidance. The following categories are available:
#### - retail and recreation, i.e. places like restaurants,cafes, shopping centers, theme parks,museums, libraries, and movie theaters
#### - grocery and pharmacy, i.e. grocery markets, food warehouses, farmers markets, specialty food shops, drug stores, and pharmacies
#### - parks, i.e. national parks, public beaches, marinas, dog parks, plazas,and public gardens
#### - transit stations i.e. all public transport hubs such as subway, bus, and train stations
#### - workplaces, i.e. places of work
#### - residential, i.e. people’s residence

#### Select a couple of European countries of your choice and analyze the trends in the previous variables over time:
#### - produce a plot of the data by averaging the observable over a period of one week ( hint: convert the data field to lubridate::week ) and one month and quantify the impact of COVID-19 restrictions on mobility sitations.

```r
# This function filters the gmr tibble for any choosen 'country', then it sums all the rows that have the same
# date value

region <- function(country)
    {
        filter(gmr, country_region == country) %>% group_by(date) %>%
        summarise(retail_and_recreation = mean(retail_and_recreation_percent_change_from_baseline),
                  grocery_and_pharmacy  = mean(grocery_and_pharmacy_percent_change_from_baseline),
                  parks                 = mean(parks_percent_change_from_baseline),
                  transit_stations      = mean(transit_stations_percent_change_from_baseline),
                  workplaces            = mean(workplaces_percent_change_from_baseline),
                  residential           = mean(residential_percent_change_from_baseline))
    }
```

```r
# This function plots all the trends for a given tibble of a specific country 
# (obtained through the function region() )

plottrends <- function(tibble)
    {
        options(repr.plot.width=15, repr.plot.height=8)

        ggplot() + 
        geom_line(data=tibble, aes(x=date, y=retail_and_recreation,color=names(tibble[2]))) + 
        geom_line(data=tibble, aes(x=date, y=grocery_and_pharmacy,color=names(tibble[3]))) + 
        geom_line(data=tibble, aes(x=date, y=parks,color=names(tibble[4]))) + 
        geom_line(data=tibble, aes(x=date, y=transit_stations,color=names(tibble[5]))) + 
        geom_line(data=tibble, aes(x=date, y=workplaces,color=names(tibble[6]))) + 
        geom_line(data=tibble, aes(x=date, y=residential,color=names(tibble[7]))) +
        labs(x = 'Week', y = 'Percent change from baseline', color = 'Places')
    }
```

```r
# We load for example the tibble regarding Italy

it_gmr <- region('Italy')
it_gmr[1:5,]
```
